```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from math import inf
```

```python
sinal_1 = np.random.normal(1,1.5,4000)
sinal_2 = np.random.normal(5,2,3000)
alarm_treshold = 2
sinal = np.concatenate([sinal_1,sinal_2])
```

```python
plt.figure(figsize=(16,4))
plt.plot(sinal)
```

```python
treshold = 2
n_1 = 4000
n_2 = 3000
x_1 = np.linspace(1,n_1,n_1)
x_2 = np.linspace(1,n_2,n_2)

kernel_1 = stats.gaussian_kde(sinal_1)
prob_1 = kernel_1.integrate_box_1d(treshold,inf)

Pa_1 = prob_1*(1-prob_1)

kernel_2 = stats.gaussian_kde(sinal_2)
prob_2 = kernel_2.integrate_box_1d(treshold,inf)

Pa_2 = prob_2*(1-prob_2)

cind_1 = sum((Pa_1/x_1)*np.exp(-Pa_1*x_1))
cind_2 = sum((Pa_2/x_2)*np.exp(-Pa_2*x_2))

print(prob_1, prob_2)
print(Pa_1, Pa_2)
print(cind_1, cind_2)
```
